```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
plt.Figure(figsize=(20,5))
plt.plot(df.index,df.Passengers,color="Green",marker='.')
plt.xlabel("Month")
plt.ylabel('Passengers')
plt.title("Airline Passengers")
plt.show()
```
