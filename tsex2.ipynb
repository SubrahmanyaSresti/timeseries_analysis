```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
```

```python
values = np.random.randn(50)
```

```python
dates= pd.date_range(start="2024-01-01", end="2024-02-19",freq="D")
df= pd.DataFrame(values,index=dates,columns=["Value"])
```

```python
df.head()
```

```python
plt.Figure(figsize=(20,5))
plt.plot(df.index,df.Value,color="teal",marker=".")
plt.xlabel("Dates")
plt.ylabel("Random values")
plt.show()
```
