```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

sns.set_theme()
```

```python
df = pd.read_csv("daily-min-temperatures.csv",parse_dates=[0],index_col="Date")
```

```python
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(20,5))
sns.lineplot(data=df)
plt.show()
```

temperature of the year 1985

```python
plt.figure(figsize=(20,5))
sns.lineplot(data=df.loc["1985-01-01":"1985-12-31"])
plt.xlabel("Dates")
plt.ylabel("Temperature in C")
plt.title("Min Temperatures of the year 1985")
```

```python
plt.figure(figsize=(20,5))
plt.title("temperature of Jan 1990")
sns.scatterplot(data=df.loc["1990-01-01":"1990-01-31"])
```

```python
pd.plotting.autocorrelation_plot(df["1990-01-01":"1990-01-31"])
```

autocorrealtion of the entire range

```python
plt.figure(figsize=(20,5))
pd.plotting.autocorrelation_plot(df)
plt.plot()
```

autocorrelation of one year

```python
pd.plotting.autocorrelation_plot(df.loc["1990-01-01":"1990-01-31"])
plt.plot()
```

```python
plt.figure(figsize=(20,5))
sns.histplot(data=df.loc["1990-01-01":"1990-12-31"],bins=100)
```

```python
df2= pd.read_csv("daily-max-temperatures.csv",parse_dates=[0],index_col="Date")
df2.head()
```

comparing minimum and maximum temperature of a month

```python
plt.figure(figsize=(20,5))
sns.kdeplot(data=df.loc["1990-01-01":"1990-01-31"],fill=True,palette="flare")
sns.kdeplot(data=df2.loc["1990-01-01":"1990-01-31"],fill=True,palette="crest")
```

```python
plt.figure(figsize=(20,5))
sns.boxplot(x=df["Temp"],palette="crest")
```

```python
sns.heatmap(df)
plt.title("heat map of all the temperatures")
plt.show()
```

```python
sns.heatmap(data=df.loc["1990-01-01":"1990-12-31"],cmap="RdPu")
plt.title("heatmap of the year 1990")
plt.show()
```

```python
sns.heatmap(data=df.loc["1990-01-01":"1990-01-31"],cmap="crest")
plt.title("Temperatures of jan 1990")
plt.show()
```
